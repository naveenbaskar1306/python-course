```python
l = [1,2,4,3,2,1,2,5,3]

t = []

for i in l:
    if i not in t:
        t.append(i)

print(t)
```

```python
l1 = ["apple", "banana", "cherry"]
l2 = [1, 2, 3]

d = {}

for i in range(len(l1)):
    d[l1[i]] = l2[i]

print(d)
```

```python
l1 = ["apple", "banana", "cherry"]
l2 = [1, 2, 3]

result = {}

for key, value in zip(l1, l2):
    result[key] = value

print(result)
```
