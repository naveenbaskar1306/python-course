```python
#Upper Case

a = "Welcome To the world"
print(a.upper())
```

```python
#Lower Case

a="WELCOME TO THE WORLD"
print(a.lower())
```

```python
#Remove Whitespace
a="   Remove Whitespace  "
print(a.strip())
```

```python
#Replace String
a="Replace String"
print(a.replace("e","n"))
```

```python
#Split String
a="Split String"
print(a.split("t"))
print(a.split())
```

```python
#capatalize
a="hello, and welcome to my world"
b=a.capitalize()
print(b)
```

```python
#casefold

a="Hello AND WElcome"
print(a.casefold())
```
