```python
Python has something called:

Integer Caching (Small Integer Optimization)

#Python automatically reuses memory for integers in the range:

-5 to 256 true 

 257 is outside the cached range.

#Python creates separate objects.

#X and Y point to different memory locations.
```

```python
1️⃣ Integers
a = 200
b = 200
print(a is b)  # True


✔ Small integers are cached.

a = 1000
b = 1000
print(a is b)  # False


❌ Large integers usually not cached.
```

```python
   Floats

a = 5.5
b = 5.5
print(a is b)  # False
Floats are not cached → different memory.

```

```python
3️⃣ String
a = "python"
b = "python"
print(a is b)  # Often True


Python interns small/simple strings.

But:

a = "python programming"
b = "python programming"
print(a is b)  # Often False


Not guaranteed.
```

```python
4️⃣ List (Mutable)
a = [1,2,3]
b = [1,2,3]
print(a is b)  # False


Lists are mutable → always new object.

But:

b = a
print(a is b)  # True


Now both reference same object.
```

```python
5️⃣ Tuple (Immutable)
a = (1,2,3)
b = (1,2,3)
print(a is b)  # Maybe True or False


May reuse memory — not guaranteed.
```

```python
6️⃣ Dictionary
a = {"x":1}
b = {"x":1}
print(a is b)  # False


Always separate objects.
```
