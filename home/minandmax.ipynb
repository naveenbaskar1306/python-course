```python
l = [1, 2, 4, 5, 6, 7,9,8,11,13,15]

minimum = l[0]
maximum = l[0]

for i in l:
    if i < minimum:
        minimum = i
    if i > maximum:
        maximum = i

print("Minimum:", minimum)
print("Maximum:", maximum)
```

```python
l = [1, 2, 4, 5, 6, 7,9,8,11,13,15]
l = list(set(l))

l.sort()

second_smallest = l[1]
second_largest = l[-2]

print("Second Smallest:", second_smallest)
print("Second Largest:", second_largest)
```
