```python
str1 = "listen"
str2 = "silent"

if sorted(str1) == sorted(str2):
    print("Anagram")
else:
    print("Not Anagram")

 #sorted() arranges letters in alphabetical order.

#"listen" → ['e','i','l','n','s','t']

#"silent" → ['e','i','l','n','s','t']
```

```python
word = "malayalam"

if word == word[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")

# [::-1] reverses the string.

#"madam" reversed → "madam"
```

```python
num = 9
flag = False

for i in range(2, num):
    if num % i == 0:
        flag = True
        break

if flag:
    print("Not Prime")
else:
    print("Prime")

#Loop checks numbers from 2 to num-1

#% means remainder.
```

```python
numbers = [4, 7, 1, 9, 3]

print("Maximum:", max(numbers))
print("Minimum:", min(numbers))

#max() → largest value
#min() → smallest value
```

```python
numbers = [4, 7, 1, 9, 3]

numbers.sort()

print("Second Largest:", numbers[-2])

```

```python
numbers = [5, 2, 8, 1, 9]

# Ascending order
numbers.sort()
print("Ascending:", numbers)

# Descending order
numbers.sort(reverse=True)
print("Descending:", numbers)
```

```python
text = input("enter the text")

vowels = "aeiou"
v_count = 0
c_count = 0

for ch in text:
    if ch in vowels:
        v_count += 1
    else:
        c_count += 1

print("Vowels:", v_count)
print("Consonants:", c_count)
```

```python
num = 1234
total = 0

for digit in str(num):
    total += int(digit)

print("Sum:", total)
```

```python
num = 5
fact = 1

for i in range(1, num+1):
    fact = fact * i

print("Factorial:", fact)
#5! = 5 × 4 × 3 × 2 × 1
   #= 120
```

```python
numbers = [1, 2, 2, 3, 3, 3, 4]

freq = {}

for num in numbers:
    if num in freq:
        freq[num] += 1
    else:
        freq[num] = 1

print(freq)
```

```python
num = 153
total = 0
for digit in str(num):
    total += int(digit) ** 3

if total == num:
    print("Armstrong Number")
else:
    print("Not Armstrong")
```
